```python
import numpy as np
import pandas as pd
np.random.seed(42)

# Generate 1000 values within the specified ranges
c1 = np.random.uniform(0.773, 1.039, 1000)
c2 = np.random.uniform(0.807, 1.067, 1000)
c3 = np.random.uniform(0.772, 1.016, 1000)
c4 = np.random.uniform(0.775, 1.027, 1000)

# Select 100 indices randomly
indices = np.random.choice(1000, 1000, replace=False)

# Set values at selected indices to be equal
c2[indices] = c1[indices]
c4[indices] = c3[indices]

# Trim arrays to have 100 values
c1 = c1[indices]
c2 = c2[indices]
c3 = c3[indices]
c4 = c4[indices]
```

```python
df = pd.DataFrame()
```

```python
df['c1']=c1
df['c2']=c2
df['c3']=c3
df['c4']=c4
```
