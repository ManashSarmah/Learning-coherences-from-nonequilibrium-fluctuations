```python
import numpy as np

# Generate some sample data from a non-normal distribution
data = df.c1.values/df.c2.values

# Set the number of bootstrap samples and the confidence level
n_bootstrap = 1000
conf_level = 0.96

# Generate the bootstrap samples
bootstrap_samples = []
for i in range(n_bootstrap):
    bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
    bootstrap_samples.append(bootstrap_sample)

# Calculate the medians of the bootstrap samples
medians = np.median(bootstrap_samples, axis=1)

# Calculate the confidence interval based on the distribution of medians
lower_ci = np.percentile(medians, 100 * (1 - conf_level) / 2)
upper_ci = np.percentile(medians, 100 * (1 + conf_level) / 2)

# Print the confidence interval
print(f"The {conf_level*100}% confidence interval for the median is: ({lower_ci:.2f}, {upper_ci:.2f})")
```
